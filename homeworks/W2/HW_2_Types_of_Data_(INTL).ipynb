## Task 2

Given a dictionary of students with their scores, find the average score of each student across all subjects and identify the student with the highest average score.

```python
students_grades = {
    "John": [85, 90, 78],
    "Alice": [88, 92, 80],
    "Bob": [75, 85, 72],
    "Diana": [90, 95, 94],
    "Charlie": [70, 65, 80]
}
```

```python
# Calculate and print the average grade for each student
student_averages = {}
for student, grades in students_grades.items():
    average_grade = sum(grades) / len(grades)
    student_averages[student] = average_grade
    print(f"{student}: Average Grade = {average_grade:.2f}")
```

```python
# Find the student with the highest average grade
highest_avg_student = None
highest_average = -1

for student, average in student_averages.items():
    if average > highest_average:
        highest_average = average
        highest_avg_student = student

print(f"Student with the highest average score: {highest_avg_student} (Average: {highest_average:.2f})")
```

# Part 2 : Working with CSV!

Add a new column to this CSV file named "Bonus," and calculate the bonus for employees in the Sales department as 10% of their MonthlyRate.

```python
# Load the CSV file
file_path = 'employee_data.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Add new column call 'Bonus'
df['Bonus'] = 0.0  # Initialize Bonus column with 0
```

```python
# Calculate 10% of bonus into the 'Bonus' Column
df.loc[df['Department'] == 'Sales', 'Bonus'] = df['MonthlyRate'] * 0.10
```

# Part 3 : Working with Matplotlib

Plot a graph showing the average salary of employees in each department to compare the average salaries across departments.

```python
# Load CSV and create DataFrame
data = pd.read_csv('employee_data.csv') # Using original data for this part
df = pd.DataFrame(data)
```

```python
# Calculate average MonthlyRate by Department
average_monthly_rate_by_department = df.groupby('Department')['MonthlyRate'].mean().reset_index()
print("Average MonthlyRate by Department:")
print(average_monthly_rate_by_department)
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_monthly_rate_by_department['Department'], average_monthly_rate_by_department['MonthlyRate'], color='skyblue')
plt.xlabel('Department')
plt.ylabel('Average Monthly Rate')
plt.title('Average Monthly Rate by Department')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
